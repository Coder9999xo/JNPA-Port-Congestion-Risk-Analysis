## 2.Temporal Analysis

##### To understand the overall trend and volatility

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values("Date").reset_index(drop=True)
df['Rolling_Mean_Dwell'] = df['Imp_Dwell_Overall'].rolling(3).mean()
df['Rolling_Std_Dwell'] = df['Imp_Dwell_Overall'].rolling(3).std()

fig, ax = plt.subplots(figsize=(14, 6))

ax.plot(df['Date'], df['Imp_Dwell_Overall'], label='Raw Data (Monthly)', color='blue', alpha=0.5, marker='o')

ax.plot(df['Date'], df['Rolling_Mean_Dwell'], label='3-Month Moving Average (Trend)', color='red', linewidth=3)

ax.fill_between(df['Date'], df['Rolling_Mean_Dwell'] - df['Rolling_Std_Dwell'], df['Rolling_Mean_Dwell'] + df['Rolling_Std_Dwell'], color='red', alpha=0.1, label='Volatility Range (Risk)')

ax.axvspan(pd.Timestamp('2023-06'), pd.Timestamp('2023-09'), color='yellow', alpha=0.15, label='Monsoon Months')
ax.axvspan(pd.Timestamp('2024-06'), pd.Timestamp('2024-09'), color='yellow', alpha=0.15)
ax.axvspan(pd.Timestamp('2025-06'), pd.Timestamp('2025-09'), color='yellow', alpha=0.15)

ax.set_xticks(df['Date'])
ax.set_xticklabels(df['Date'].dt.strftime('%b-%Y'), rotation=90)

ax.set_title("Standard Time Series Analysis: Trend vs. Volatility", fontsize=16, weight='bold')
ax.set_ylabel("Import Dwell Time (Hours)")
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout() 
plt.show()
```

1. The width of the volatility band (Pink Shadow) expands significantly over time. In 2023, the variance was contained. By 2025, the volatility band widens drastically during Q3, indicating that the port's ability to buffer against disruptions has degraded.
2. In 2023, the port handled seasonal stress easily dwell times less than 25 hrs. By 2025, the same triggers caused massive spikes greater than 43 hrs. This suggests the system breaks once volume exceeds.
3. The Blue Line frequently crosses the upper limit of the Moving Average (Red Line) during crisis months. This proves that relying on "Monthly Averages" or "Trends" is insufficient for Demurrage mitigation.

##### Shock in previous month effecting current month  
Traffic spikes don’t always break the port right away. Instead, those extra containers pile up in the yard, creating a delayed bottleneck where the "cleanup" pressure only starts spiking dwell times a month later.
To test this hypothesis, the relationship between the previous month's 
throughput and the current month's import dwell time is examined.

```python
df["Volume_Lag1"] = df["Total_TEUs_Handled"].shift(1)
fig, ax1 = plt.subplots(figsize=(10,6))

line1 = ax1.plot(df["Date"],df["Imp_Dwell_Overall"],color="blue",marker="o",label="Import Dwell (t)")
ax1.set_ylabel("Import Dwell (Hours)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()

line2 = ax2.plot(df["Date"],df["Total_TEUs_Handled"],color="orange",label="Volume (t)")

line3 = ax2.plot(df["Date"],df["Volume_Lag1"],color="green",linestyle="--",label="Volume (t-1)")

ax2.set_ylabel("Container Volume (TEUs)", color="orange")
ax2.tick_params(axis="y", labelcolor="orange")

lines = line1 + line2 + line3
labels = [l.get_label() for l in lines]

ax1.legend(lines, labels, loc="upper left")


plt.title("Temporal Relationship: Throughput and Import Dwell")

fig.tight_layout()

plt.show()
corr_value = df["Volume_Lag1"].corr(df["Imp_Dwell_Overall"])
print("Correlation:", corr_value)
```

The data suggests a lagged congestion effect, where a surge in volume often leads to higher dwell times a month later (correlation ≈ 0.49). It’s not a 1:1 match every time, but it points to a logistical pile-up—containers stack up in the yard, and that pressure eventually comes down to evacuation process in the following weeks.

## 3.Bivariate Analysis

##### Objective: To mathematically confirm which input variables (X) have the strongest predictive power over the target variables (Y). We use a Correlation Matrix to identify relationships and Scatter Plots to visualize the linearity of these drivers.

##### Correlation Heatmap : To identify strongest decision drivers.

```python
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(14, 10))
heatmap = sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
heatmap.set_title('Correlation Matrix', fontsize=16, weight='bold')
plt.show()
```

1. Plot A Train vs Overall: It is an inconsistent relationship.Trend is positive,low overall dwell times still coexist with high train dwell times.This shows that train bottlenecks are independent of port congestion.

## 4.Multivariate Analysis

##### Monsoon Effect Check:
##### Assumption:Rain acts as a Force Multiplier, not just an addition. High Volume is manageable in dry weather. Rain is manageable with low volume. But High Volume + Rain causes an Exponential (Non-Linear) collapse.

```python

df['Season'] = df['Date'].dt.month.isin([6, 7, 8, 9]).map({True: 'Monsoon', False: 'Dry'})

plt.figure(figsize=(12, 8))

sns.scatterplot(data=df, x='Total_TEUs_Handled', y='Imp_Dwell_Overall', hue='Season',style='Season',palette={'Monsoon': 'red', 'Dry': 'blue'},s=150,alpha=0.7)

plt.axvline(x=600000, color='grey', linestyle='--', label='Capacity Threshold (~600k)')

plt.title("Checking if rain acts as dwell time multiplier", fontsize=16, weight='bold')
plt.xlabel("Port Volume (TEUs)")
plt.ylabel("Overall Dwell Time (Hours)")
plt.legend(title="Seasonality")
plt.grid(True, alpha=0.3)

plt.show()
```

1. To the left of dashed line which is less than 600k TEUs Monsoon and Dry months are almost clustered together in 15-25 hr range so with low volume monsoon months can be still handled.
2. Towards right of this dashed line the clustering is less and there are more spread out, even at high volumes the dry months are manageable with less than 30 hrs, but monsoon months are shoting upto high times like 35 to 40 and above hrs.Indicating volume and rain have a non linear relationship.
3. To tackle this problem of high volume in addition with monsoon a new feature must be engineered Stress=Volume*Is_Monsoon.

## Feature Engineering Plan:


#### 1.Rail_Friction:
We observed from Temporal analysis truck vs train that truck lines flat (stable) while rail lines spiked at times proving they operate at different stress levels.
So a ratio directly showing how much slower rail is compared to road helps in separating the rail network failure from regular port delay.

##### Rail_Friction=(Imp_Dwell_Train / Imp_Dwell_Truck)

#### 2.Stress:
 We observed from Multivariate Analysis graph that Red (Monsoon) month data points were significantly high commpared to Blue (Dry) month data points only when volume exceeded 600k TEUs.To capture this non-linear congestion which happens when there is interaction between high volume and monsoon so we engineer Stress.

##### Stress=(Total_TEUs_Handled * Is_Monsoon)


#### 3.Volume_Lag1

The temporal analysis indicates that increases in container throughput are sometimes followed by higher import dwell times in the subsequent month, suggesting a delayed congestion effect. To quantify this relationship, the correlation between previous-month throughput and current import dwell time was computed, yielding r ≈ 0.49, which represents a moderate positive association.

a lagged throughput feature was engineered to capture this delayed effect:
##### Volume_Lag1=Total TEUs Handled(t−1)
	​
